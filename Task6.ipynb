```python
# Smart Parking Lot Management System

def space_available(in_count, capacity):
    if in_count == capacity:
        return "Not Available"
    else:
        return "Available"

def smart_parking_lot(capacity, vehicle_logs):
    in_count = 0
    out_count = 0
    for vehicle in vehicle_logs:
        if vehicle == 'IN':
            if in_count == capacity:
                status = space_available(in_count, capacity)
                print("Parking Full")
                print(f"Parking Status: {status}")
                print("Wait for vehicles to move out")
                print("--" * 20)
            else:
                in_count += 1
                print("Vehicle Entered")
        else:
            print("Vehicle Exited")
            out_count += 1
            in_count -= 1
        status = space_available(in_count, capacity)
        print(f"Currently Parked Vehicles: {in_count}")
        print(f"Parking Status: {status}")
        print("--" * 20)

parking_capacity = int(input("Parking Capacity: "))
vehicles = []
entry = True
print("Are there vehicles to add in the logs?")
while(entry):
    choice = input("Enter 'IN' or 'OUT' to add it to the logs or enter 'no' to exit: ").upper()
    if choice == 'NO':
        entry = False
    elif choice == 'IN' or choice == 'OUT':
        vehicles.append(choice)
    else:
        print("Enter correct choice")
smart_parking_lot(parking_capacity, vehicles)
    
```

```python
# Online Food Delivery Time Estimator

def calculate_eta(distance, traffic, weather):
    base_speed = 40 
    eta = (distance / base_speed) * 60
    if traffic == "HIGH":
        eta += 20
    elif traffic == "MEDIUM":
        eta += 10
    else:
        eta += 0
    if weather == "RAINY":
        eta += 15
    elif weather == "STORMY":
        eta += 25
    else:
        eta += 0
    return eta

def delivery_system():
    distance = float(input("Distance (km): "))
    traffic = input("Traffic Level (Low/Medium/High): ").upper()
    weather = input("Weather (Clear/Rainy/Stormy): ").upper()
    estimated_time = calculate_eta(distance, traffic, weather)
    print("--" * 20)
    print(f"Estimated Delivery Time: {int(estimated_time)} minutes")

delivery_system()
```

```python
# Movie Theatre Seat Occupancy Analyzer

def analyze_booking(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100
    print("--" * 20)
    print(f"Total Seats: {total_seats}")
    print(f"Booked Seats: {booked_count}")
    print(f"Occupancy: {int(occupancy)}%")
    if occupancy == 100:
        print("Show Status: Housefull")
        print("Suggestion: Open additional show")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Monitor bookings closely")
    elif occupancy >= 50:
        print("Show Status: Filling Fast")
    else:
        print("Show Status: Seats Available")

def booking_system():
    total = int(input("Total Seats: "))
    booked_number = int(input("Enter number of booked seats: "))
    booked = []
    for i in range(booked_number):
        booked.append(1)
    analyze_booking(total, booked)

booking_system()
```

```python
# Cloud Server Load Classification System

def analyze_server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading
    average = total / len(cpu_readings)
    print("--" * 20)
    print(f"CPU Readings: {cpu_readings}")
    print(f"Average CPU Load: {int(average)}%")
    if average < 50:
        print("Server Status: Normal")
    elif average >= 50 and average < 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

def server_monitoring_system():
    readings = []
    print("Enter CPU readings. Type -1 to stop.")
    while True:
        value = int(input("CPU Usage (%): "))
        if value == -1:
            break
        else:
            readings.append(value)
    if len(readings) > 0:
        analyze_server_load(readings)
    else:
        print("No CPU readings provided.")

server_monitoring_system()
```

```python
# Smart Classroom Resource Usage Monitor 

def analyze_resource_usage(resources):
    overused = []
    limit = 8
    print("--" * 20)
    print("Resource Usage Report")
    for resource, hours in resources.items():
        print(f"{resource}: {hours} hours")
        if hours > limit:
            overused.append(resource)
    print("--" * 20)
    if len(overused) > 0:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

def classroom_system():
    resources = {}
    print("Enter classroom resource usage (in hours)")
    resources["Projector"] = int(input("Projector usage hours: "))
    resources["AC"] = int(input("AC usage hours: "))
    resources["Lights"] = int(input("Lights usage hours: "))
    analyze_resource_usage(resources)

classroom_system()
```

```python
# Online Event Registration Capacity Controller

def manage_registrations(capacity, total_requests):
    confirmed = 0
    waitlist = 0
    for i in range(total_requests):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlist += 1
    print("--" * 20)
    print(f"Event Capacity: {capacity}")
    print(f"Total Registration Requests: {total_requests}")
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlist}")
    if confirmed == capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")

def registration_system():
    capacity = int(input("Event Capacity: "))
    registrations = int(input("Number of Registration Requests: "))
    manage_registrations(capacity, registrations)

registration_system()
```
